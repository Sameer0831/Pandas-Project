# 16. Replacing salary values (<=50k and > 50k) with 0 and 1

```python
def salary_mod(sal):
    if sal == '<=50K':
        return 0
    else:
        return 1
```

```python
data['encoded_salary'] = data['income'].apply(salary_mod)
```

```python
data.head()
```

# 17. Which Workclass getting highest salary

```python
data.groupby('workclass')[['encoded_salary']].mean().sort_values(by='encoded_salary', ascending=False)
```

# 18. Who Has better chance to get salary >50k male or female

```python
data.groupby('gender')[['encoded_salary']].count()
```
